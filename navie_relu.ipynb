```python
def navie_relu(x) :
    assert len(x.shape) == 2
    x = x.copy()
    for i in range (x.shape[0]):
        for j in range(x.shape[1]):
            x[i,j] = max(x[i,j], 0)
    return x
```

```python
def ndot(x,y):
    assert len(x.shape) ==1
    assert len(y.shape) ==1
    assert x.shape[0] == y.shape[0]
    z = 0
    for i in range(x.shape[0]):
        z += x[i] * y[i]
    return z
```

```python
x = np.array(range(9))
y = np.array(range(9))
z = ndot(x , y)
z
```
